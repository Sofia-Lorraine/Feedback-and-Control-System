```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1] #numerator
den = [1, 2, 1]     #denominator s^2 + 2s + 2
G = ctrl.TransferFunction(num, den)
print("Transfer Function G(s):")
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles of G(s):", poles)
```

```python
zero = ctrl.zeros(G)
print("Zeros of G(s):")
print(zero)
```

```python
import matplotlib.pyplot as plt

ctrl.pzmap(G, grid=True, title='Pole-Zero Map of G(s)')
plt.show()
```
