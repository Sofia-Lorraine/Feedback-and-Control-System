```python
import serial
import time
import sys

SERIAL_PORT = "COM3"   # Change if needed
BAUD_RATE = 9600 

# ---- Connect to Arduino ----
try:
    ser = serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=1)
    time.sleep(2)  # Wait for Arduino reset
    print(f"Connected to Arduino on {SERIAL_PORT}")

except serial.SerialException as e:
    print(f"Error: Could not open serial port: {e}")
    sys.exit()


def main():

    print("\n--- Arduino Servo Controller ---")
    print("Enter 0–180 or 'q' to quit.\n")

    try:
        while True:

            user_input = input("Enter degrees: ").strip()

            if user_input.lower() == 'q':
                print("Program ended.")
                break

            # Validate input
            if not user_input.isdigit():
                print("Error: Numbers only.")
                continue

            angle = int(user_input)

            if angle < 0 or angle > 180:
                print("Error: Range must be 0–180.")
                continue

            # Send to Arduino
            ser.write(f"{angle}\n".encode('utf-8'))

            time.sleep(0.1)

            # Read response
            if ser.in_waiting > 0:
                feedback = ser.readline().decode('utf-8', errors='ignore').strip()
                print("Arduino:", feedback)

    except KeyboardInterrupt:
        print("\nInterrupted by user.")

    finally:
        ser.close()
        print("Serial connection closed.")


if __name__ == "__main__":
    main()
```
