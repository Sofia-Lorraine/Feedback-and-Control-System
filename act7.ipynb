```python
import serial
import time

# --- CONFIGURATION ---
# Replace 'COM3' with your actual Arduino port (e.g., '/dev/ttyUSB0' on Linux/Mac)
SERIAL_PORT = 'COM3' 
BAUD_RATE = 9600

def main():
    try:
        # Initialize serial communication
        # timeout=1 ensures the read blocking won't hang forever
        arduino = serial.Serial(port=SERIAL_PORT, baudrate=BAUD_RATE, timeout=1)
        
        # Arduino resets when serial opens; wait a moment for it to boot up
        print("Connecting to Arduino...")
        time.sleep(2) 
        
        print("Connected! Type an angle (0-180) and press Enter. Type 'exit' to quit.")

        while True:
            # Get user input from the Python console
            user_input = input("Enter angle: ").strip()

            if user_input.lower() == 'exit':
                print("Exiting program.")
                break

            if not user_input:
                continue

            # Send the angle followed by a newline character (\n) as required by your Arduino code
            # We encode it to bytes because serial communication transfers bytes, not strings
            arduino.write(f"{user_input}\n".encode('utf-8'))

            # Wait a brief moment for Arduino to process and respond
            time.sleep(0.1)

            # Read the response from Arduino
            if arduino.in_waiting > 0:
                # Read the line, decode from bytes to string, and strip whitespace
                response = arduino.readline().decode('utf-8').strip()
                print(f"Arduino says: {response}")
                
    except serial.SerialException as e:
        print(f"Serial Error: {e}")
    except KeyboardInterrupt:
        print("\nProgram interrupted by user.")
    finally:
        # Ensure the connection is closed properly when exiting
        if 'arduino' in locals() and arduino.is_open:
            arduino.close()
            print("Serial connection closed.")

if __name__ == "__main__":
    main()
```
