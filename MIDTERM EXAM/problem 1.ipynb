```python
import control
import matplotlib.pyplot as plt
import numpy as np

# System Equations:
# f(t) = (D^2 + 2D + 2)x1 - (D + 1)x2
# 0 = (D + 1)x1 - (D^2 + 2D + 2)x2

def main():
    # Define coefficients
    num = [1, 1]
    den = [1, 4, 7, 6, 3]

    # Create transfer function
    sys = control.TransferFunction(num, den)
    
    # 1. Step Response Analysis
    t, y = control.step_response(sys)
    plt.figure(figsize=(10, 6))
    plt.plot(t, y, linewidth=2, color='#2563eb')
    plt.title('Step Response')
    plt.grid(True)
    plt.savefig('step_response.png')

    # 2. Pole-Zero Mapping
    plt.figure(figsize=(10, 6))
    control.pzmap(sys, plot=True)
    plt.title('Pole-Zero Map')
    plt.grid(True)
    plt.savefig('pole_zero_map.png')

if __name__ == "__main__":
    main()
    
```
