```python

import control
import matplotlib.pyplot as plt

# System Definition
num = [6]
den = [16, 24, 52, 36, 0]
sys = control.TransferFunction(num, den)
```

# Transient Analysis (Step Response)
t, y = control.step_response(sys)
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.title("Step Response of the Mechanical System")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.savefig('step_response.png')
