```python
# Define the transfer function H(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
import control
import matplotlib.pyplot as plt
import numpy as np

# Numerator: 1 | Denominator: s^4 + 2s^3 + 3s^2 + 2s + 0
num = [1]
den = [1, 2, 3, 2, 0]
sys = control.TransferFunction(num, den)

# Generate Step Response
t, y = control.step_response(sys)

plt.figure(figsize=(10, 6))
plt.plot(t, y, linewidth=2, color='#0284c7')
plt.title('System Step Response')
plt.xlabel('Time (seconds)')
plt.ylabel('Outcome (theta_2)')
plt.grid(True, alpha=0.3)
plt.savefig('step_response.png')

# Generate Pole-Zero Map
plt.figure(figsize=(8, 8))
control.pzmap(sys, plot=True)
plt.title('Pole-Zero Map')
plt.grid(True, alpha=0.3)
plt.savefig('pole_zero_map.png')

plt.show()
```
